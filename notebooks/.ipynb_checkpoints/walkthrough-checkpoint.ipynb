### The forecast is a decline curve — *"expected"* is what it emits
Per-well **Arps hyperbolic** parameters are drawn; the full-uptime potential is written as the `Forecast` series in
`PRODUCT_VOLUME_SUMMARY`, and actual oil in `WELL_VOL_DAILY` applies a performance factor and the on-stream fraction
(`generator.py:238`). That forecast is the baseline for every variance, deferment, and decline KPI (ADR 0006). Let's plot
one well's actual vs expected oil — the gap is the surveillance signal.

````python
from oag_generator import schema

t = canonical_table_paths(DS)
con = duckdb.connect()
WELL_ID = 1   # UWI "NO 15/9-F-1" — the deterministically seeded worst-actor / trap well

actual = con.execute(f"""
    SELECT VOLUME_DATE AS d, OIL_VOLUME AS oil
    FROM read_parquet('{t["well_vol_daily"]}')
    WHERE WELL_ID = {WELL_ID} ORDER BY d
""").fetchdf()

forecast = con.execute(f"""
    SELECT p.START_DATE AS d, p.VOLUME AS oil
    FROM read_parquet('{t["product_volume_summary"]}') p
    JOIN read_parquet('{t["reporting_entity"]}') re USING (REPORTING_ENTITY_ID)
    WHERE re.REPORTING_ENTITY_KIND = '{schema.KIND_WELL}' AND re.ASSOCIATED_OBJECT_ID = {WELL_ID}
      AND p.PRODUCT = '{schema.PRODUCT_OIL}' AND p.QUANTITY_METHOD = '{schema.QUANTITY_FORECAST}'
    ORDER BY d
""").fetchdf()

import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(pd.to_datetime(forecast.d), forecast.oil, label="expected (Arps forecast)", lw=2.2, color="#C67A1E")
ax.plot(pd.to_datetime(actual.d), actual.oil, label="actual oil", lw=1.6, color="#2F8479")
ax.fill_between(pd.to_datetime(actual.d), actual.oil, forecast.oil, color="#C67A1E", alpha=.10)
ax.set_title(f"Well {WELL_ID} (NO 15/9-F-1) — actual vs expected oil (bbl/day)")
ax.set_ylabel("bbl / day"); ax.legend(); ax.grid(alpha=.25); fig.tight_layout()
plt.show()
````

### From physical columns to the semantic model

Each of the nine physical tables above is annotated by an **OSI semantic model** — authored once in
[`semantic/semantic_models.yaml`](../semantic/semantic_models.yaml) (ADR 0008/0011). The semantic layer holds no data; it
holds *governed metadata* over these same OSDU columns, so tools stop arguing about which column means what. For each table it
declares:

- **entities** — the model's grain (`type: primary`) and its join keys (`type: foreign`), each mapped to a physical column;
- **dimensions** — the columns you group/filter by, typed **`time`** or **`categorical`**;
- **measures** — the aggregatable quantities (`agg: sum`/`count`/`average`) that KPIs are built from.

This is technically Layer 3, but it's built directly on the Layer-2 columns, so we meet it here. `manifest.py` reads the YAML
into typed objects (`Entity`/`Dimension`/`Measure`/`SemanticModel`); `load_semantic_layer()` resolves the path itself, so it
works regardless of the notebook's working directory.

````python
from oag_semantic.manifest import SEMANTIC_DIR, load_semantic_layer
REPO = SEMANTIC_DIR.parent   # repo root, derived from the package — cwd-independent

layer = load_semantic_layer()
print("semantic models (one per canonical table):", list(layer.models))
````

#### The primary key, sourced in three coordinated places

The primary key you saw promoted into the OSDU JSON `id` (`oag:WELL:1`) is declared and consumed in three places that must
agree — all referencing the **same OSDU column names**. A subtlety worth stating: the vendored profile
`spec/osdu/pdm_profile.json` pins column **names + types** (the conformance authority) but does **not** itself declare which
column is the key — the PK *designation* is the project's, following PDM's surrogate-key convention, and lives in `schema.py`
(physical) and the semantic YAML (semantic grain).

````python
# 1. OSDU profile — the authority for column NAMES + TYPES (no PK designation here)
prof_well = json.loads((REPO / "spec/osdu/pdm_profile.json").read_text())["tables"]["well"]
print("1. pdm_profile.json  -> columns:",
      [c["name"] for c in prof_well["columns"]][:4], "...  (names+types only)")

# 2. schema.py — the PHYSICAL primary key (drives the Parquet + the OSDU JSON `id`)
print("2. schema.WELL       -> primary_key:", schema.WELL.primary_key)

# 3. semantic_models.yaml — the SEMANTIC grain (entity of type 'primary')
well_model = layer.model("well")
pk_entity = next(e for e in well_model.entities if e.type == "primary")
print("3. semantic 'well'   -> primary entity:", (pk_entity.name, pk_entity.type, pk_entity.expr))
print("\nall three point at the SAME OSDU column: WELL_ID")
````

#### Anatomy of one semantic model

Take `well_vol_daily` — the richest model. Its `entities` name the grain and the well join key; its one `time` dimension is the
axis every window filters on; its `measures` are the governed aggregations the surveillance/deferment/rollup KPIs draw from
(`actual_oil_volume` is a `sum` of the physical `OIL_VOLUME` column).

````python
def describe_model(name):
    m = layer.model(name)
    print(f"model {m.name!r}  ->  canonical table (node_relation.alias): {m.table!r}\n")
    ent = df([{"name": e.name, "type": e.type, "expr (physical col)": e.expr} for e in m.entities])
    dim = df([{"name": d.name, "type": d.type, "expr (physical col)": d.expr,
               "granularity": d.time_granularity} for d in m.dimensions])
    mea = df([{"name": me.name, "agg": me.agg, "expr (physical col)": me.expr} for me in m.measures])
    return ent, dim, mea

entities, dimensions, measures = describe_model("well_vol_daily")
print("ENTITIES (grain + join keys):");    display(entities)
print("DIMENSIONS (group/filter axes):");  display(dimensions)
print("MEASURES (aggregations for KPIs):"); display(measures)
````

#### `name` vs `type` vs `expr` — the distinction that trips people up

Look at the forecast time dimension: `name: forecast_date`, `type: time`, `expr: START_DATE`. These answer three *different*
questions:

- **`name`** is a **semantic alias you invent** — there is no `forecast_date` column. It gives the column business meaning.
  The same physical `START_DATE` is named differently per model (`forecast_date` here, `event_date` on downtime,
  `alloc_start_date` on allocation) because it means something different in each.
- **`type: time`** is a **modeling role** (the model's time axis), *not* a storage datatype. In fact `START_DATE` is stored as
  a plain **string** (ISO-8601) — the compile filters it with `WHERE START_DATE BETWEEN ? AND ?` on ISO strings, which sort
  lexicographically, so no date casting is needed.
- **`expr`** is the **physical column** (or any SQL expression) the alias resolves to. It's the token dropped into the SQL.

````python
pvs = layer.model("product_volume_summary")
tdim = pvs.time_dimension()
print("semantic time dimension :", tdim)
print("  name (alias)          :", tdim.name, "  <- not a real column")
print("  type (semantic role)  :", tdim.type)
print("  expr (physical column):", tdim.expr)
print("  physical storage type :", schema.PRODUCT_VOLUME_SUMMARY.arrow_schema().field(tdim.expr).type,
      "  <- a string, NOT a date")

print("\nsame physical column, different semantic names per model:")
df([{"model": name, "expr": "START_DATE",
     "semantic name": next(d.name for d in layer.model(name).dimensions if d.expr == "START_DATE")}
    for name in ("product_volume_summary", "down_time_event", "rpen_allocation_factor")])
````

#### How it's authored, read, and kept honest

The manifest is **hand-authored**, not generated from the data — governed definitions written once (ADR 0008). It's *read* at
runtime by `load_semantic_layer()` (plain PyYAML, no heavy deps on the answer path), *validated* dev-time by MetricFlow
(`validate_manifest()`, run in §4), and kept **OSDU-conformant** by `referenced_columns()`: it collects every physical `expr`
per table, and a conformance test asserts each one is a real column in `pdm_profile.json` — so a measure can never quietly
reference a column that drifted away from the OSDU profile.

````python
referenced = layer.referenced_columns()   # {table -> set of physical columns the manifest touches}
print("columns the manifest references on 'well':", sorted(referenced["well"]))

# the conformance check, in miniature: every referenced column must exist physically
physical = set(schema.WELL.column_names)
drifted = referenced["well"] - physical
print("physical WELL columns              :", sorted(physical))
print("referenced columns not in physical :", drifted or "none — conformant ✓")
````

````python
# term resolution — NL business phrase -> governed metric + comparison + threshold
for phrase in ["below expected", "underperforming", "below forecast", "watering out"]:
    concept = lpg.resolve_term(phrase)
    if concept:
        print(f"{phrase:<16} -> metric {concept.metric!r}  ({concept.comparison} {concept.threshold_ref})")
    else:
        print(f"{phrase:<16} -> (no governed term; round-1 vocab ships the surveillance signal only)")
````

## 8 · Axis-B — the assessment harness  <code>src/oag_harness</code>

The harness scores competing assistant implementations on the seven-dimension rubric. Governing principle (ADR 0015):
**every dimension has an objective, computed anchor**; the assessor panel is only a tiebreaker; effort is *reported, never
scored*. We exercise each anchor below.

### Dimension 1 — functional correctness (the oracle scores 100%)
`submission_from_gold` builds the answer an *oracle* implementation would return; `score_submissions` grades the whole set
(`functional.py:241`). An unanswered gradable question **fails** — you can't inflate the pass rate by answering only the sure
ones (ADR 0025).

````python
from oag_harness.functional import SPECS, submission_from_gold, score_submissions

oracle = {}
for q in catalog.questions():
    spec, gold_path = SPECS.get(q.gold_id), BIG / q.gold_artifact
    if spec is None or not gold_path.exists():
        continue
    oracle[q.id] = submission_from_gold(json.loads(gold_path.read_text()), spec, q.id, q.expected_behavior)

score = score_submissions(oracle, BIG)
print(f"graded {score.n_graded} questions -> pass_rate {score.pass_rate:.0%}  ({score.n_correct} correct)")
print(score.summary())
````

### Dimension 1, hardened — the held-out evaluation seed (anti-cheat)
The gold shipped in the fork is *build-time collateral*, not the graded number. At round close the dataset is regenerated at
an **unseen seed** and gold recomputed (`evalseed.py:191`, ADR 0016). Watch the *same oracle answers* — built from the
fork's seed — collapse from 100% to a partial score on the held-out seed. Only the behavior-only (ambiguous/trap) questions
survive; every hard-coded numeric answer fails. A seed-agnostic implementation would recompute and pass.

````python
from oag_harness.evalseed import grade_on_eval_seed

EVAL_SEED = 8675309   # a seed the "contestant" never saw
run = grade_on_eval_seed(oracle, big_cfg, EVAL_SEED, WORK / "eval")
print(f"fork-time answers on the held-out seed -> pass_rate {run.score.pass_rate:.0%} "
      f"(was 100% on the shipped gold)")
print("\npublished, reproducible record:")
run.published()
````

### Dimension 7 — sealed custody + locus adherence
Round-2 change requests are authored **before** the fork tag and held out of version control; only their content-addressed

### Dimension 4 — test quality (seeded-bug perturbation probe)
Does a contestant's *own* suite catch a planted defect? The probe checks the suite is green, perturbs a gold value, re-runs,
then always restores (`probes.py:37`). `caught=True` means the suite noticed.

````python
from oag_harness.probes import run_perturbation_probe

baseline = json.loads((BIG / "gold" / "surveillance.json").read_text())
state = {"gold": json.loads(json.dumps(baseline))}
def suite_passes(): return state["gold"]["flagged"] == baseline["flagged"]   # stand-in for a real suite
def perturb():      state["gold"]["flagged"][0]["shortfall_bbl"] += 999.0
def restore():      state["gold"] = json.loads(json.dumps(baseline))

probe = run_perturbation_probe("bump a gold shortfall by 999 bbl", suite_passes, perturb, restore)
print("baseline green :", probe.baseline_green)
print("caught the bug :", probe.caught, "| conclusive:", probe.conclusive)
````
